```python
# # Advanced Feature Engineering for Maximum Accuracy
# print("üîß Creating advanced features...")

# # Calculate Ahmedabad city center (approximate)
# ahmedabad_center_lat = 23.0225
# ahmedabad_center_lon = 72.5714

# # 1. Distance-based features
# df['distance_from_center'] = np.sqrt(
#     (df['lat'] - ahmedabad_center_lat)**2 + 
#     (df['lon'] - ahmedabad_center_lon)**2
# )

# # 2. Luxury score based on area and amenities
# df['luxury_score'] = (
#     (df['area_sqft'] / 1000) * 0.3 +  # Area contribution
#     df['bathrooms'] * 0.2 +            # Bathroom contribution
#     (df['bedrooms'] / 5) * 0.2 +       # Bedroom ratio
#     (1 / (df['distance_from_center'] + 0.01)) * 0.3  # Location premium
# )

# # 3. Price per sqft
# df['price_per_sqft'] = df['price_inr'] / df['area_sqft']

# # 4. Room density
# df['room_density'] = (df['bedrooms'] + df['bathrooms']) / df['area_sqft'] * 1000

# # 5. Location clusters (based on lat/lon)
# from sklearn.cluster import KMeans
# kmeans = KMeans(n_clusters=8, random_state=42, n_init=10)
# df['location_cluster'] = kmeans.fit_predict(df[['lat', 'lon']])

# # 6. Interaction features
# df['bedrooms_x_bathrooms'] = df['bedrooms'] * df['bathrooms']
# df['area_x_luxury'] = df['area_sqft'] * df['luxury_score']

# # 7. Binned features
# df['area_bin'] = pd.cut(df['area_sqft'], bins=10, labels=False)
# df['price_bin'] = pd.cut(df['price_inr'], bins=10, labels=False)

# # 8. Statistical features by neighborhood
# neighborhood_stats = df.groupby('locality')['price_inr'].agg(['mean', 'std', 'median']).reset_index()
# neighborhood_stats.columns = ['locality', 'locality_price_mean', 'locality_price_std', 'locality_price_median']
# df = df.merge(neighborhood_stats, on='locality', how='left')

# # Fill any NaN values
# df['locality_price_std'] = df['locality_price_std'].fillna(df['locality_price_std'].mean())

# print(f"‚úÖ Enhanced features created! New shape: {df.shape}")
# print(f"üìä Feature columns: {list(df.columns)}")

# Advanced Feature Engineering for Maximum Accuracy (FIXED)
print("üîß Creating advanced features...")

# Initialize processor without file path
processor = RealDataProcessor()

# Load the raw CSV data
df_raw = pd.read_csv('../data/ahmedabad.csv')
print(f"üìä Raw data loaded: {len(df_raw)} rows")

# Process the data using the correct method
df = processor.process_real_ahmedabad_data(df_raw)

print(f"‚úÖ Data processed: {len(df)} properties")
print(f"üìè Features: {df.shape[1]} columns")
print(f"üí∞ Price range: ‚Çπ{df['price'].min():,.0f} - ‚Çπ{df['price'].max():,.0f}")

# Calculate Ahmedabad city center (approximate)
ahmedabad_center_lat = 23.0225
ahmedabad_center_lon = 72.5714

# Check if we have latitude/longitude columns
if 'latitude' in df.columns and 'longitude' in df.columns:
    lat_col, lon_col = 'latitude', 'longitude'
elif 'lat' in df.columns and 'lon' in df.columns:
    lat_col, lon_col = 'lat', 'lon'
else:
    # Create dummy coordinates for Ahmedabad neighborhoods
    print("‚ö†Ô∏è No coordinates found, creating estimated coordinates...")
    np.random.seed(42)
    df['latitude'] = np.random.normal(ahmedabad_center_lat, 0.05, len(df))
    df['longitude'] = np.random.normal(ahmedabad_center_lon, 0.05, len(df))
    lat_col, lon_col = 'latitude', 'longitude'

# 1. Distance-based features
df['distance_from_center'] = np.sqrt(
    (df[lat_col] - ahmedabad_center_lat)**2 + 
    (df[lon_col] - ahmedabad_center_lon)**2
)

# 2. Luxury score based on available features
df['luxury_score'] = (
    (df['square_feet'] / 1000) * 0.3 +  # Area contribution
    df['bathrooms'] * 0.2 +             # Bathroom contribution
    (df['bedrooms'] / 5) * 0.2 +        # Bedroom ratio
    (1 / (df['distance_from_center'] + 0.01)) * 0.3  # Location premium
)

# 3. Price per sqft
df['price_per_sqft'] = df['price'] / df['square_feet']

# 4. Room density
df['room_density'] = (df['bedrooms'] + df['bathrooms']) / df['square_feet'] * 1000

# 5. Location clusters (based on lat/lon)
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=8, random_state=42, n_init=10)
df['location_cluster'] = kmeans.fit_predict(df[[lat_col, lon_col]])

# 6. Interaction features
df['bedrooms_x_bathrooms'] = df['bedrooms'] * df['bathrooms']
df['area_x_luxury'] = df['square_feet'] * df['luxury_score']

# 7. Binned features
df['area_bin'] = pd.cut(df['square_feet'], bins=10, labels=False)
df['price_bin'] = pd.cut(df['price'], bins=10, labels=False)

# 8. Statistical features by neighborhood
if 'neighborhood' in df.columns:
    neighborhood_stats = df.groupby('neighborhood')['price'].agg(['mean', 'std', 'median']).reset_index()
    neighborhood_stats.columns = ['neighborhood', 'neighborhood_price_mean', 'neighborhood_price_std', 'neighborhood_price_median']
    df = df.merge(neighborhood_stats, on='neighborhood', how='left')
    
    # Fill neighborhood stats NaN values with appropriate defaults
    df['neighborhood_price_std'] = df['neighborhood_price_std'].fillna(df['neighborhood_price_std'].mean())
    df['neighborhood_price_mean'] = df['neighborhood_price_mean'].fillna(df['price'].mean())
    df['neighborhood_price_median'] = df['neighborhood_price_median'].fillna(df['price'].median())

# 9. Property age features (if year_built is available)
if 'property_age' in df.columns:
    # Create age categories with proper handling
    df['age_category'] = pd.cut(df['property_age'], 
                               bins=[0, 5, 10, 20, 50], 
                               labels=['New', 'Recent', 'Moderate', 'Old'],
                               include_lowest=True)
    
    # Fill NaN values in age_category with 'Unknown' category
    df['age_category'] = df['age_category'].cat.add_categories(['Unknown'])
    df['age_category'] = df['age_category'].fillna('Unknown')

# 10. Additional derived features
df['room_efficiency'] = df['bedrooms'] / df['square_feet'] * 1000
df['bathroom_ratio'] = df['bathrooms'] / df['bedrooms'].replace(0, 1)  # Avoid division by zero

# Handle NaN values carefully for each column type
print("üßπ Cleaning NaN values...")

# For numeric columns, fill with appropriate values
numeric_columns = df.select_dtypes(include=[np.number]).columns
for col in numeric_columns:
    if col != 'price':  # Don't fill target variable
        df[col] = df[col].fillna(df[col].median())

# For categorical columns, handle separately
categorical_columns = df.select_dtypes(include=['category', 'object']).columns
for col in categorical_columns:
    if col in df.columns:
        if df[col].dtype.name == 'category':
            # Add 'Unknown' category if not exists and fill NaN
            if 'Unknown' not in df[col].cat.categories:
                df[col] = df[col].cat.add_categories(['Unknown'])
            df[col] = df[col].fillna('Unknown')
        else:
            # For object columns
            df[col] = df[col].fillna('Unknown')

print(f"‚úÖ Enhanced features created! New shape: {df.shape}")
print(f"üìä All columns: {list(df.columns)}")

# Show sample of enhanced data
print("\nüîç Sample Enhanced Data:")
print(df[['price', 'bedrooms', 'bathrooms', 'square_feet', 'luxury_score', 'distance_from_center']].head())
```

```python
# # Prepare Data for Training (90% train, 10% test)
# print("üéØ Preparing data for maximum accuracy training...")

# # Select features (exclude target and non-numeric)
# exclude_cols = ['price_inr', 'locality', 'property_type', 'url']
# feature_cols = [col for col in df.columns if col not in exclude_cols]

# # Handle categorical variables
# categorical_cols = df[feature_cols].select_dtypes(include=['object']).columns
# if len(categorical_cols) > 0:
#     print(f"üè∑Ô∏è Encoding categorical columns: {list(categorical_cols)}")
#     df_encoded = pd.get_dummies(df[feature_cols], columns=categorical_cols, drop_first=True)
# else:
#     df_encoded = df[feature_cols]

# X = df_encoded
# y = df['price_inr']

# print(f"üìä Final feature matrix shape: {X.shape}")
# print(f"üéØ Target variable shape: {y.shape}")

# # Split data: 90% train, 10% test for maximum accuracy
# X_train, X_test, y_train, y_test = train_test_split(
#     X, y, test_size=0.1, random_state=42, shuffle=True
# )

# print(f"üöÇ Training set: {X_train.shape[0]} samples")
# print(f"üß™ Test set: {X_test.shape[0]} samples")
# print(f"üìà Training ratio: {X_train.shape[0] / len(df) * 100:.1f}%")

# # Scale features
# scaler = RobustScaler()
# X_train_scaled = scaler.fit_transform(X_train)
# X_test_scaled = scaler.transform(X_test)

# print("‚úÖ Data preparation complete!")


# Prepare Data for Training (90% train, 10% test)
print("üéØ Preparing data for maximum accuracy training...")

# Select features (exclude target and non-numeric)
exclude_cols = ['price']  # Main target variable

# Add other columns to exclude if they exist
for col in ['neighborhood', 'property_type', 'description']:
    if col in df.columns:
        exclude_cols.append(col)

feature_cols = [col for col in df.columns if col not in exclude_cols]

# Handle categorical variables
df_features = df[feature_cols].copy()
categorical_cols = df_features.select_dtypes(include=['object']).columns

if len(categorical_cols) > 0:
    print(f"üè∑Ô∏è Encoding categorical columns: {list(categorical_cols)}")
    df_encoded = pd.get_dummies(df_features, columns=categorical_cols, drop_first=True)
else:
    df_encoded = df_features

# Remove any remaining non-numeric columns
df_encoded = df_encoded.select_dtypes(include=[np.number])

X = df_encoded
y = df['price']

print(f"üìä Final feature matrix shape: {X.shape}")
print(f"üéØ Target variable shape: {y.shape}")
print(f"üí∞ Target range: ‚Çπ{y.min():,.0f} - ‚Çπ{y.max():,.0f}")

# Split data: 90% train, 10% test for maximum accuracy
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42, shuffle=True
)

print(f"üöÇ Training set: {X_train.shape[0]} samples")
print(f"üß™ Test set: {X_test.shape[0]} samples")
print(f"üìà Training ratio: {X_train.shape[0] / len(df) * 100:.1f}%")

# Scale features
scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("‚úÖ Data preparation complete!")
```

```python
# Memory-Efficient Model Training Function
def train_model_safe(model, X_train, y_train, model_name, cv_folds=3):
    """Train model with memory management and error handling"""
    print(f"üîß Training {model_name}...")
    
    try:
        # Fit the model
        model.fit(X_train, y_train)
        
        # Cross-validation with reduced folds
        cv_scores = cross_val_score(model, X_train, y_train, cv=cv_folds, 
                                   scoring='r2', n_jobs=1)  # Single job to prevent crashes
        
        # Clean up memory
        gc.collect()
        
        return {
            'model': model,
            'cv_mean': cv_scores.mean(),
            'cv_std': cv_scores.std(),
            'success': True
        }
    except Exception as e:
        print(f"‚ùå Error training {model_name}: {str(e)}")
        return {
            'model': None,
            'cv_mean': 0,
            'cv_std': 0,
            'success': False
        }

print("‚úÖ Memory-efficient training function ready!")
```

```python
# Train Multiple Models (Sequential for Memory Safety)
print("üöÄ Starting sequential model training...")

models_to_train = {
    'Random Forest': RandomForestRegressor(
        n_estimators=200, max_depth=15, min_samples_split=5,
        min_samples_leaf=2, random_state=42, n_jobs=1
    ),
    'Gradient Boosting': GradientBoostingRegressor(
        n_estimators=200, max_depth=8, learning_rate=0.1,
        min_samples_split=5, random_state=42
    ),
    'Extra Trees': ExtraTreesRegressor(
        n_estimators=150, max_depth=12, min_samples_split=5,
        min_samples_leaf=2, random_state=42, n_jobs=1
    ),
    'Ridge Regression': Ridge(alpha=1.0),
    'Elastic Net': ElasticNet(alpha=0.1, l1_ratio=0.5, random_state=42)
}

# Add XGBoost if available
if XGBOOST_AVAILABLE:
    models_to_train['XGBoost'] = xgb.XGBRegressor(
        n_estimators=200, max_depth=8, learning_rate=0.1,
        random_state=42, n_jobs=1
    )

# Add LightGBM if available
if LIGHTGBM_AVAILABLE:
    models_to_train['LightGBM'] = lgb.LGBMRegressor(
        n_estimators=200, max_depth=8, learning_rate=0.1,
        random_state=42, n_jobs=1, verbose=-1
    )

# Train models sequentially
trained_models = {}
results = []

for name, model in models_to_train.items():
    # Use scaled data for linear models, original for tree-based
    if name in ['Ridge Regression', 'Elastic Net']:
        train_X = X_train_scaled
    else:
        train_X = X_train
    
    result = train_model_safe(model, train_X, y_train, name)
    
    if result['success']:
        trained_models[name] = {
            'model': result['model'],
            'use_scaled': name in ['Ridge Regression', 'Elastic Net']
        }
        
        results.append({
            'Model': name,
            'CV R¬≤ Mean': f"{result['cv_mean']:.4f}",
            'CV R¬≤ Std': f"{result['cv_std']:.4f}"
        })
    
    # Clean memory after each model
    gc.collect()

# Display results
results_df = pd.DataFrame(results)
print("\nüìä Cross-Validation Results:")
print(results_df.to_string(index=False))

print(f"\n‚úÖ Successfully trained {len(trained_models)} models!")
```

```python
# Evaluate Models on Test Set
print("üß™ Evaluating models on test set...")

test_results = []

for name, model_info in trained_models.items():
    model = model_info['model']
    use_scaled = model_info['use_scaled']
    
    # Use appropriate test data
    test_X = X_test_scaled if use_scaled else X_test
    
    # Make predictions
    y_pred = model.predict(test_X)
    
    # Calculate metrics
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
    
    test_results.append({
        'Model': name,
        'R¬≤ Score': f"{r2:.4f}",
        'MAE (‚Çπ)': f"{mae:,.0f}",
        'RMSE (‚Çπ)': f"{rmse:,.0f}",
        'MAPE (%)': f"{mape:.1f}%"
    })

# Display results
test_results_df = pd.DataFrame(test_results)
test_results_df = test_results_df.sort_values('R¬≤ Score', ascending=False)

print("\nüèÜ Test Set Performance (Sorted by R¬≤ Score):")
print(test_results_df.to_string(index=False))

# Find best model
best_model_name = test_results_df.iloc[0]['Model']
best_r2 = float(test_results_df.iloc[0]['R¬≤ Score'])

print(f"\nü•á Best Model: {best_model_name} (R¬≤ = {best_r2:.4f})")

if best_r2 > 0.72:
    print("üéâ SUCCESS! Accuracy improved beyond target!")
else:
    print(f"üìà Current best: {best_r2:.4f} (Target: >0.72)")
```

```python
# Create Ensemble Model for Maximum Accuracy
print("ü§ù Creating ensemble model...")

# Select top 3 performing models for ensemble
top_models = []
for i in range(min(3, len(test_results_df))):
    model_name = test_results_df.iloc[i]['Model']
    if model_name in trained_models:
        model_info = trained_models[model_name]
        top_models.append((model_name, model_info['model']))

if len(top_models) >= 2:
    # Create voting regressor
    ensemble = VotingRegressor(
        estimators=top_models,
        n_jobs=1  # Single job for stability
    )
    
    print(f"üîß Training ensemble with models: {[name for name, _ in top_models]}")
    
    # Train ensemble (use original features since tree models dominate)
    ensemble.fit(X_train, y_train)
    
    # Evaluate ensemble
    ensemble_pred = ensemble.predict(X_test)
    ensemble_r2 = r2_score(y_test, ensemble_pred)
    ensemble_mae = mean_absolute_error(y_test, ensemble_pred)
    ensemble_rmse = np.sqrt(mean_squared_error(y_test, ensemble_pred))
    ensemble_mape = np.mean(np.abs((y_test - ensemble_pred) / y_test)) * 100
    
    print(f"\nü§ù Ensemble Model Performance:")
    print(f"   R¬≤ Score: {ensemble_r2:.4f}")
    print(f"   MAE: ‚Çπ{ensemble_mae:,.0f}")
    print(f"   RMSE: ‚Çπ{ensemble_rmse:,.0f}")
    print(f"   MAPE: {ensemble_mape:.1f}%")
    
    # Check if ensemble is better
    if ensemble_r2 > best_r2:
        print(f"\nüéØ Ensemble improved accuracy: {ensemble_r2:.4f} > {best_r2:.4f}")
        best_model_name = 'Ensemble'
        best_r2 = ensemble_r2
        trained_models['Ensemble'] = {'model': ensemble, 'use_scaled': False}
    else:
        print(f"\nüìä Individual model still better: {best_r2:.4f} > {ensemble_r2:.4f}")
else:
    print("‚ö†Ô∏è Not enough models for ensemble")

print(f"\nüèÜ Final Best Model: {best_model_name} (R¬≤ = {best_r2:.4f})")
```

```python
# Feature Importance Analysis
print("üìä Analyzing feature importance...")

# Get the best model
best_model = trained_models[best_model_name]['model']

# Extract feature importance (works for tree-based models)
if hasattr(best_model, 'feature_importances_'):
    importances = best_model.feature_importances_
    feature_names = X.columns
elif hasattr(best_model, 'estimators_'):  # For ensemble
    # Average importance across ensemble models
    importances = np.mean([est.feature_importances_ for est in best_model.estimators_ 
                          if hasattr(est, 'feature_importances_')], axis=0)
    feature_names = X.columns
else:
    print("‚ö†Ô∏è Feature importance not available for this model type")
    importances = None

if importances is not None:
    # Create feature importance dataframe
    feature_imp_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': importances
    }).sort_values('Importance', ascending=False)
    
    print("\nüîù Top 15 Most Important Features:")
    print(feature_imp_df.head(15).to_string(index=False))
    
    # Plot feature importance
    plt.figure(figsize=(12, 8))
    top_features = feature_imp_df.head(15)
    plt.barh(range(len(top_features)), top_features['Importance'])
    plt.yticks(range(len(top_features)), top_features['Feature'])
    plt.xlabel('Feature Importance')
    plt.title(f'Top 15 Feature Importances - {best_model_name}')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()

# Memory cleanup
gc.collect()
print("\n‚úÖ Feature importance analysis complete!")
```

```python
# Test Case 4: Cross-Validation Robustness Testing
print("\nüß™ TEST CASE 4: Cross-Validation Robustness")
print("=" * 45)

from sklearn.model_selection import cross_val_score, KFold, StratifiedKFold
from sklearn.model_selection import cross_validate

try:
    # Perform k-fold cross validation
    print("üîÑ Performing 5-Fold Cross-Validation...")
    
    kfold = KFold(n_splits=5, shuffle=True, random_state=42)
    
    # Multiple scoring metrics
    scoring = ['r2', 'neg_mean_absolute_error', 'neg_root_mean_squared_error']
    cv_results = cross_validate(gb_model, X_train, y_train, cv=kfold, 
                               scoring=scoring, return_train_score=True)
    
    print(f"\nüìà Cross-Validation Results:")
    
    # R¬≤ scores
    r2_scores = cv_results['test_r2']
    print(f"   R¬≤ Score - Mean: {r2_scores.mean():.4f} (¬±{r2_scores.std()*2:.4f})")
    print(f"   R¬≤ Score - Range: {r2_scores.min():.4f} to {r2_scores.max():.4f}")
    
    # MAE scores (convert back from negative)
    mae_scores = -cv_results['test_neg_mean_absolute_error']
    print(f"   MAE - Mean: ‚Çπ{mae_scores.mean():,.0f} (¬±{mae_scores.std()*2:,.0f})")
    print(f"   MAE - Range: ‚Çπ{mae_scores.min():,.0f} to ‚Çπ{mae_scores.max():,.0f}")
    
    # RMSE scores (convert back from negative)
    rmse_scores = -cv_results['test_neg_root_mean_squared_error']
    print(f"   RMSE - Mean: ‚Çπ{rmse_scores.mean():,.0f} (¬±{rmse_scores.std()*2:,.0f})")
    print(f"   RMSE - Range: ‚Çπ{rmse_scores.min():,.0f} to ‚Çπ{rmse_scores.max():,.0f}")
    
    # Consistency check
    r2_consistency = (r2_scores.max() - r2_scores.min()) < 0.05
    print(f"\n‚úÖ Model Consistency Check:")
    if r2_consistency:
        print("   üéØ CONSISTENT: R¬≤ variation < 0.05 across folds")
    else:
        print("   ‚ö†Ô∏è VARIABLE: R¬≤ shows significant variation across folds")
    
    # Plot cross-validation results
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 3, 1)
    plt.boxplot(r2_scores)
    plt.title('R¬≤ Score Distribution')
    plt.ylabel('R¬≤ Score')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 3, 2)
    plt.boxplot(mae_scores)
    plt.title('MAE Distribution')
    plt.ylabel('MAE (‚Çπ)')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 3, 3)
    plt.boxplot(rmse_scores)
    plt.title('RMSE Distribution')
    plt.ylabel('RMSE (‚Çπ)')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
except Exception as e:
    print(f"‚ùå Error in test case 4: {e}")
```

```python
# Test Case 6: Robustness Testing (Missing Data & Outliers)
print("\nüß™ TEST CASE 6: Robustness Testing")
print("=" * 40)

try:
    # 1. Test with missing data simulation
    print("üîç Testing robustness to missing data...")
    
    # Create test data with some missing values
    X_test_missing = X_test.copy()
    
    # Randomly set 5% of values to NaN in non-critical columns
    np.random.seed(42)
    for col in X_test_missing.columns:
        if col not in ['bedrooms', 'square_feet', 'bathrooms']:  # Keep critical features
            mask = np.random.random(len(X_test_missing)) < 0.05
            X_test_missing.loc[mask, col] = np.nan
    
    # Fill NaN values with median (robust imputation)
    X_test_imputed = X_test_missing.fillna(X_test_missing.median())
    
    # Make predictions with imputed data
    y_pred_imputed = gb_model.predict(X_test_imputed)
    
    # Compare with original predictions
    original_pred = gb_model.predict(X_test)
    imputation_diff = np.abs(y_pred_imputed - original_pred)
    mean_diff = imputation_diff.mean()
    max_diff = imputation_diff.max()
    
    print(f"   Mean prediction difference: ‚Çπ{mean_diff:,.0f}")
    print(f"   Max prediction difference: ‚Çπ{max_diff:,.0f}")
    print(f"   Correlation with original: {np.corrcoef(original_pred, y_pred_imputed)[0,1]:.4f}")
    
    if mean_diff < y_test.mean() * 0.05:  # Less than 5% of average price
        print("   ‚úÖ ROBUST: Model handles missing data well")
    else:
        print("   ‚ö†Ô∏è SENSITIVE: Model significantly affected by missing data")
    
    # 2. Test with outlier detection
    print(f"\nüéØ Testing outlier predictions...")
    
    # Identify outliers in predictions
    residuals = np.abs(y_test - original_pred)
    Q1 = np.percentile(residuals, 25)
    Q3 = np.percentile(residuals, 75)
    IQR = Q3 - Q1
    outlier_threshold = Q3 + 1.5 * IQR
    
    outlier_indices = np.where(residuals > outlier_threshold)[0]
    outlier_count = len(outlier_indices)
    outlier_percentage = (outlier_count / len(y_test)) * 100
    
    print(f"   Outlier predictions: {outlier_count} ({outlier_percentage:.1f}%)")
    print(f"   Outlier threshold: ‚Çπ{outlier_threshold:,.0f}")
    
    if outlier_percentage < 5:
        print("   ‚úÖ LOW OUTLIERS: <5% outlier predictions")
    elif outlier_percentage < 10:
        print("   ‚ö†Ô∏è MODERATE OUTLIERS: 5-10% outlier predictions")
    else:
        print("   ‚ùå HIGH OUTLIERS: >10% outlier predictions")
    
    # Analyze worst predictions
    if outlier_count > 0:
        print(f"\nüîç Analyzing worst predictions:")
        worst_indices = outlier_indices[:min(5, outlier_count)]  # Top 5 worst
        
        for i, idx in enumerate(worst_indices):
            actual = y_test.iloc[idx]
            predicted = original_pred[idx]
            error = abs(actual - predicted)
            error_pct = (error / actual) * 100
            
            print(f"   #{i+1}: Actual ‚Çπ{actual:,.0f}, Predicted ‚Çπ{predicted:,.0f}, Error {error_pct:.1f}%")
    
    # 3. Test model stability with noisy data
    print(f"\nüåä Testing stability with noisy data...")
    
    # Add small random noise to test data
    noise_level = 0.01  # 1% noise
    X_test_noisy = X_test.copy()
    
    for col in X_test_noisy.select_dtypes(include=[np.number]).columns:
        noise = np.random.normal(0, X_test_noisy[col].std() * noise_level, len(X_test_noisy))
        X_test_noisy[col] = X_test_noisy[col] + noise
    
    y_pred_noisy = gb_model.predict(X_test_noisy)
    
    # Compare stability
    noise_diff = np.abs(y_pred_noisy - original_pred)
    noise_correlation = np.corrcoef(original_pred, y_pred_noisy)[0,1]
    
    print(f"   Mean difference with noise: ‚Çπ{noise_diff.mean():,.0f}")
    print(f"   Correlation with original: {noise_correlation:.4f}")
    
    if noise_correlation > 0.95:
        print("   ‚úÖ STABLE: Model robust to small data perturbations")
    elif noise_correlation > 0.90:
        print("   ‚ö†Ô∏è MODERATE: Model somewhat sensitive to noise")
    else:
        print("   ‚ùå UNSTABLE: Model highly sensitive to noise")
    
except Exception as e:
    print(f"‚ùå Error in test case 6: {e}")
```

```python
# Test Case 7: Performance Analysis by Price Segments
print("\nüß™ TEST CASE 7: Price Segment Analysis")
print("=" * 45)

try:
    # Divide test data into price segments
    print("üìä Analyzing performance across price segments...")
    
    # Create price quartiles
    price_quartiles = pd.qcut(y_test, q=4, labels=['Low', 'Medium-Low', 'Medium-High', 'High'])
    y_pred_test = gb_model.predict(X_test)
    
    # Analyze performance by price segment
    segment_analysis = []
    
    for segment in ['Low', 'Medium-Low', 'Medium-High', 'High']:
        mask = price_quartiles == segment
        if mask.sum() > 0:
            y_true_segment = y_test[mask]
            y_pred_segment = y_pred_test[mask]
            
            r2_segment = r2_score(y_true_segment, y_pred_segment)
            mae_segment = mean_absolute_error(y_true_segment, y_pred_segment)
            mape_segment = np.mean(np.abs((y_true_segment - y_pred_segment) / y_true_segment)) * 100
            
            segment_analysis.append({
                'Segment': segment,
                'Count': mask.sum(),
                'Price Range': f"‚Çπ{y_true_segment.min():,.0f} - ‚Çπ{y_true_segment.max():,.0f}",
                'R¬≤ Score': f"{r2_segment:.4f}",
                'MAE': f"‚Çπ{mae_segment:,.0f}",
                'MAPE': f"{mape_segment:.1f}%"
            })
    
    segment_df = pd.DataFrame(segment_analysis)
    print("\nüìà Performance by Price Segment:")
    print(segment_df.to_string(index=False))
    
    # Visualize segment performance
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    segments = ['Low', 'Medium-Low', 'Medium-High', 'High']
    colors = ['blue', 'green', 'orange', 'red']
    
    for i, segment in enumerate(segments):
        mask = price_quartiles == segment
        if mask.sum() > 0:
            y_true_seg = y_test[mask]
            y_pred_seg = y_pred_test[mask]
            
            # Scatter plot for each segment
            row, col = i // 2, i % 2
            axes[row, col].scatter(y_true_seg, y_pred_seg, alpha=0.6, color=colors[i])
            axes[row, col].plot([y_true_seg.min(), y_true_seg.max()], 
                               [y_true_seg.min(), y_true_seg.max()], 'k--', lw=1)
            axes[row, col].set_xlabel('Actual Price (‚Çπ)')
            axes[row, col].set_ylabel('Predicted Price (‚Çπ)')
            axes[row, col].set_title(f'{segment} Price Segment')
            axes[row, col].grid(True, alpha=0.3)
            
            # Add R¬≤ score
            r2_seg = r2_score(y_true_seg, y_pred_seg)
            axes[row, col].text(0.05, 0.95, f'R¬≤ = {r2_seg:.3f}', 
                               transform=axes[row, col].transAxes,
                               bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    plt.tight_layout()
    plt.show()
    
    # Performance consistency analysis
    r2_scores = [float(row['R¬≤ Score']) for row in segment_analysis]
    r2_range = max(r2_scores) - min(r2_scores)
    
    print(f"\n‚úÖ Segment Analysis Results:")
    print(f"   R¬≤ Score Range: {r2_range:.4f}")
    
    if r2_range < 0.1:
        print("   üéØ CONSISTENT: Similar performance across price segments")
    elif r2_range < 0.2:
        print("   ‚ö†Ô∏è MODERATE: Some variation across price segments")
    else:
        print("   ‚ùå INCONSISTENT: Significant variation across price segments")
    
    # Identify best and worst performing segments
    best_segment = segment_analysis[np.argmax(r2_scores)]['Segment']
    worst_segment = segment_analysis[np.argmin(r2_scores)]['Segment']
    
    print(f"   Best performing segment: {best_segment} (R¬≤ = {max(r2_scores):.4f})")
    print(f"   Worst performing segment: {worst_segment} (R¬≤ = {min(r2_scores):.4f})")
    
except Exception as e:
    print(f"‚ùå Error in test case 7: {e}")
```
