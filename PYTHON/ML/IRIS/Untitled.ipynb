```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv(r"C:\Users\dipes\Desktop\JAVA\CODE\PYTHON\ML\IRIS\Iris.csv")
```

```python
df.head()

df.tail()

df.isnull()
```

```python
# 1. Histogram 
# df.hist(figsize=(12, 8))
df.hist(figsize=(12,8))
plt.suptitle('Distribution of Iris Features')
plt.show()
```

```python
# 2. Pair plot 
sb.pairplot(df, hue='Species', diag_kind='hist')
plt.show()
```

```python
# 2. Correlation Heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = df.select_dtypes(include=[np.number]).corr()
sb.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Iris Dataset - Feature Correlation Heatmap')
plt.show()
```
