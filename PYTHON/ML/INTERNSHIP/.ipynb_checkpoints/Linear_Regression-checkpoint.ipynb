```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#loading the dataset
diabetes = load_diabetes()
X = diabetes.data  # inputs
y = diabetes.target  # outputs
```

```python
#feature scaling so that mean and std deviation is 0  for all features

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Training Set (80%) → Used to teach the model.
#testing set : 20% -> to test if it learnt 

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.1, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
# # Plot (corrected)
# plt.figure(figsize=(8, 6))
# plt.scatter(X_test.ravel(), y_test, color='red', label='Actual Points')  # Flatten X_test
# plt.plot(X_test.ravel(), y_pred, color='blue', linewidth=2, label='Predicted Line')  # Flatten X_test
# plt.title("Linear Regression on BMI vs Diabetes Progression")
# plt.xlabel("BMI (Standardized)")
# plt.ylabel("Diabetes Progression")
# plt.legend()
# plt.grid(True)
# plt.show()


# # Use only the BMI feature (e.g., column index 2)
# bmi_index = 2  # Change this if your BMI feature is at another index
# X_bmi_test = X_test[:, bmi_index]

# # Plot
# plt.figure(figsize=(6, 6))
# plt.scatter(X_bmi_test, y_test, color='red', label='Actual Points')
# plt.plot(X_bmi_test, y_pred, color='blue', linewidth=2, label='Predicted Line')  # y_pred must match X_bmi_test
# plt.title("Linear Regression on BMI vs Diabetes Progression")
# plt.xlabel("BMI")
# plt.ylabel("Disease Progression")
# plt.legend()
# plt.show()


# Assume your model is called 'model' and BMI is at index 2
bmi_index = 2


# Extract only the BMI feature for both train and test sets
X_train_bmi = X_train[:, bmi_index].reshape(-1, 1)
X_test_bmi = X_test[:, bmi_index].reshape(-1, 1)

# Train a new model using only BMI (if not already done)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_bmi, y_train)

# Predict
y_pred = model.predict(X_test_bmi)

# Plot
plt.figure(figsize=(6, 6))
plt.scatter(X_test_bmi, y_test, color='red', label='Actual Points')
plt.plot(X_test_bmi, y_pred, color='blue', linewidth=2, label='Predicted Line')
plt.title("Linear Regression on BMI vs Diabetes Progression")
plt.xlabel("BMI")
plt.ylabel("Disease Progression")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, predictions)
print("Mean Squared Error:", mse)


print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)

# Print R^2 score
r2 = model.score(X_test_bmi, y_test)
print(f"R² score: {r2:.4f}")
```
