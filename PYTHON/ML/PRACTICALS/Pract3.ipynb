## Handle outliers

### Subtask:
Detect and handle outliers in daily case counts.


**Reasoning**:
Convert the 'Date' column to datetime objects and calculate daily new cases by grouping by state and differencing confirmed cases. Handle negative daily cases by replacing them with 0.



```python
df_encoded['Date'] = pd.to_datetime(df_encoded['Date'])
df_encoded['Daily_Cases'] = df_encoded.groupby('State/UnionTerritory')['Confirmed'].diff().fillna(0)
df_encoded['Daily_Cases'] = df_encoded['Daily_Cases'].apply(lambda x: max(0, x))
display(df_encoded[['Date', 'State/UnionTerritory_Andhra Pradesh', 'Daily_Cases']].head())
```

**Reasoning**:
The previous code failed because 'State/UnionTerritory' column was dropped during one-hot encoding. I need to use the original dataframe 'df' to calculate daily cases per state before merging it with the encoded dataframe 'df_encoded'.



```python
df['Date'] = pd.to_datetime(df['Date'])
daily_cases_per_state = df.groupby('State/UnionTerritory').apply(lambda x: x.set_index('Date')['Confirmed'].diff().fillna(0)).reset_index()
daily_cases_per_state.rename(columns={'Confirmed': 'Daily_Cases'}, inplace=True)
daily_cases_per_state['Daily_Cases'] = daily_cases_per_state['Daily_Cases'].apply(lambda x: max(0, x))

df_merged = pd.merge(df, daily_cases_per_state, on=['State/UnionTerritory', 'Date'], how='left')

Q1 = df_merged['Daily_Cases'].quantile(0.25)
Q3 = df_merged['Daily_Cases'].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR

df_merged['Daily_Cases'] = df_merged['Daily_Cases'].apply(lambda x: min(x, upper_bound))

df_encoded = pd.get_dummies(df_merged, columns=['State/UnionTerritory'], drop_first=True)

display(df_encoded[['Date', 'State/UnionTerritory_Andhra Pradesh', 'Daily_Cases']].head())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
daily_cases_per_state = df.groupby('State/UnionTerritory').apply(lambda x: x.set_index('Date')['Confirmed'].diff().fillna(0)).reset_index()
daily_cases_per_state.rename(columns={'Confirmed': 'Daily_Cases'}, inplace=True)
daily_cases_per_state['Daily_Cases'] = daily_cases_per_state['Daily_Cases'].apply(lambda x: max(0, x))

df_merged = pd.merge(df, daily_cases_per_state, on=['State/UnionTerritory', 'Date'], how='left')

Q1 = df_merged['Daily_Cases'].quantile(0.25)
Q3 = df_merged['Daily_Cases'].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR

df_merged['Daily_Cases'] = df_merged['Daily_Cases'].apply(lambda x: min(x, upper_bound))

df_encoded = pd.get_dummies(df_merged, columns=['State/UnionTerritory'], drop_first=True)

display(df_encoded[['Date', 'State/UnionTerritory_Andhra Pradesh', 'Daily_Cases']].head())
```

## Scale numerical features

### Subtask:
Scale numerical features using a chosen scaler.


```python
from sklearn.preprocessing import StandardScaler

numerical_cols = ['Sno', 'ConfirmedIndianNational', 'ConfirmedForeignNational', 'Cured', 'Deaths', 'Confirmed', 'Daily_Cases']
scaler = StandardScaler()

df_encoded[numerical_cols] = scaler.fit_transform(df_encoded[numerical_cols])

display(df_encoded.head())
```

**Reasoning**:
The error message indicates that there are non-numeric values ('-') in the columns selected for scaling. These need to be converted to a numerical type, likely by replacing '-' with 0 or NaN, before scaling. Given the context of the data, replacing '-' with 0 for numerical columns seems appropriate.



```python
df_encoded.replace('-', 0, inplace=True)

numerical_cols = ['Sno', 'ConfirmedIndianNational', 'ConfirmedForeignNational', 'Cured', 'Deaths', 'Confirmed', 'Daily_Cases']
for col in numerical_cols:
    df_encoded[col] = pd.to_numeric(df_encoded[col])

scaler = StandardScaler()
df_encoded[numerical_cols] = scaler.fit_transform(df_encoded[numerical_cols])

display(df_encoded.head())
```

## Generate visualizations

### Subtask:
Create visualizations to highlight trends and patterns.


**Reasoning**:
Create visualizations as requested in the instructions to highlight trends and patterns. This involves creating a line plot for daily cases in selected states, a bar plot for total cases in the top states, and a heatmap for the correlation matrix of numerical features.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Line plot for Daily_Cases over time for selected states
selected_states = ['Maharashtra', 'Delhi', 'Kerala']
plt.figure(figsize=(15, 7))
for state in selected_states:
    state_data = df_merged[df_merged['State/UnionTerritory'] == state]
    plt.plot(state_data['Date'], state_data['Daily_Cases'], label=state)

plt.title('Daily Cases Trend for Selected States')
plt.xlabel('Date')
plt.ylabel('Daily Cases')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Bar plot for total cases for the top 10 states
top_10_states = df_merged.groupby('State/UnionTerritory')['Confirmed'].max().sort_values(ascending=False).head(10).index.tolist()
top_10_data = df_merged[df_merged['State/UnionTerritory'].isin(top_10_states)]

# Get the latest entry for each state to represent total cases
latest_top_10_data = top_10_data.loc[top_10_data.groupby('State/UnionTerritory')['Date'].idxmax()]

latest_top_10_data = latest_top_10_data.sort_values(by='Confirmed', ascending=False)

plt.figure(figsize=(15, 7))
bar_width = 0.25
r1 = range(len(latest_top_10_data))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

plt.bar(r1, latest_top_10_data['Confirmed'], color='blue', width=bar_width, edgecolor='grey', label='Confirmed')
plt.bar(r2, latest_top_10_data['Cured'], color='green', width=bar_width, edgecolor='grey', label='Cured')
plt.bar(r3, latest_top_10_data['Deaths'], color='red', width=bar_width, edgecolor='grey', label='Deaths')

plt.xlabel('State/Union Territory', fontweight='bold')
plt.ylabel('Number of Cases', fontweight='bold')
plt.title('Total Confirmed, Cured, and Deaths for Top 10 States')
plt.xticks([r + bar_width for r in range(len(latest_top_10_data))], latest_top_10_data['State/UnionTerritory'], rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()

# 3. Heatmap for correlation matrix of numerical features in df_encoded
numerical_cols_encoded = ['Sno', 'ConfirmedIndianNational', 'ConfirmedForeignNational', 'Cured', 'Deaths', 'Confirmed', 'Daily_Cases']
correlation_matrix = df_encoded[numerical_cols_encoded].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.show()
```

## Perform correlation analysis

### Subtask:
Examine relationships between variables using correlation analysis.


**Reasoning**:
Analyze the correlation matrix to identify strong positive and negative correlations and summarize the findings.



```python
print("Correlation Matrix:")
print(correlation_matrix)

print("\nAnalysis of Correlation Matrix:")

# Identify strong positive correlations (e.g., correlation coefficient > 0.7)
strong_positive_correlations = correlation_matrix.unstack().sort_values(kind="quicksort", ascending=False)
strong_positive_correlations = strong_positive_correlations[strong_positive_correlations > 0.7]
strong_positive_correlations = strong_positive_correlations[strong_positive_correlations != 1.0] # Exclude self-correlation
print("\nStrong Positive Correlations (> 0.7):")
print(strong_positive_correlations)

# Identify strong negative correlations (e.g., correlation coefficient < -0.7)
strong_negative_correlations = correlation_matrix.unstack().sort_values(kind="quicksort", ascending=True)
strong_negative_correlations = strong_negative_correlations[strong_negative_correlations < -0.7]
print("\nStrong Negative Correlations (< -0.7):")
print(strong_negative_correlations)

# Summarize key findings
print("\nSummary of Key Findings:")
print("- 'Confirmed', 'Cured', and 'Deaths' show very strong positive correlations with each other (all > 0.9). This is expected as these represent cumulative counts related to the pandemic's progression.")
print("- 'Daily_Cases' shows a moderate positive correlation with 'Confirmed', 'Cured', and 'Deaths' (around 0.4 to 0.49). This indicates that higher daily cases contribute to the increase in cumulative numbers.")
print("- 'Sno' has moderate positive correlations with 'Confirmed', 'Cured', and 'Deaths' (around 0.3 to 0.4), reflecting the cumulative nature of these metrics over time.")
print("- 'ConfirmedIndianNational' and 'ConfirmedForeignNational' have weak or negligible correlations with the other numerical features, which might be due to the changing reporting methods or their smaller scale compared to overall confirmed cases.")
print("- There are no strong negative correlations among the numerical features.")
```
