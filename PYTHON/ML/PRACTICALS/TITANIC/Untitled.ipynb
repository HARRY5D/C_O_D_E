```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv(r"C:\Users\dipes\Desktop\JAVA\CODE\PYTHON\ML\TITANIC\Titanic-Dataset.csv")
```

```python
df.head()
```

```python
df.hist(figsize=(12, 8))

plt.show()
```

```python
# 2. Survival Count Bar Chart
plt.figure(figsize=(8, 6))
survival_counts = df['Survived'].value_counts()
plt.bar(['Did not survive', 'Survived'], survival_counts.values, color=['red', 'green'])
plt.title('Survival Count')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# 6. Fare Distribution Box Plot
plt.figure(figsize=(5, 6))
sb.boxplot(data=df, x='Survived', y='Fare')
plt.title('Fare Distribution by Survival Status')
plt.xlabel('Survival Status (0: Did not survive, 1: Survived)')
plt.ylabel('Fare')
plt.show()
```
